# **Image to Pencil Sketch with Python:**

We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. 

# **Importing** **libraries**

```python
import cv2
from google.colab.patches import cv2_imshow
```

# **Reading of Image**

```python
image = cv2.imread('HP.png')
```

```python
print("Original Image")
cv2_imshow(image)
```

```python
#Reading the b/w image
bw_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print("Gray Image")
cv2_imshow(bw_image)
```

```python
#inverting the bw image
inverted_image = 255-bw_image
cv2_imshow(inverted_image)
```

#**Blurring** **the** **image**

```python

blurr_image = cv2.GaussianBlur(inverted_image,(21,21),0)
cv2_imshow(blurr_image)
```

```python
#inverting blurred image
inv_blur_image = 255-blurr_image
cv2_imshow(inv_blur_image)
```

```python
pencil_sketch = cv2.divide(bw_image,inv_blur_image,scale=256.0)
cv2_imshow(pencil_sketch)
```
